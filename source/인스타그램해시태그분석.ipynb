### 크롤링 함수

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd

from selenium.webdriver.common.keys import Keys
import getpass
import re
```

```python
def infinite_scroll(wd):
    page_height = wd.execute_script("return document.body.scrollHeight")

    print("SCROLLING START")
    
    while True:
        wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        sleep(1)
        scroll_height = wd.execute_script("return document.body.scrollHeight")

        if page_height == scroll_height:
            break
        page_height = scroll_height
    
    print("SCROLLING FINISHED")
    
    return
```

```python
def login_instagram(wd):
    url = 'https://www.instagram.com'
    wd.get(url)
    print("OPEN PAGE : " + url)
    
    while True:
        user = input("아이디를 입력세요 : ")
        input_user = wd.find_element_by_name('username')

        input_user.send_keys(Keys.CONTROL + "a")
        input_user.send_keys(Keys.DELETE)
        input_user.send_keys(user)

        password = getpass.getpass("비밀번호를 입력하세요 : ")
        input_pw = wd.find_element_by_name('password')

        input_pw.send_keys(Keys.CONTROL + "a")
        input_pw.send_keys(Keys.DELETE)
        input_pw.send_keys(password)

        input_pw.submit()
        
        sleep(2)
        
        is_login = wd.find_elements_by_class_name('Fifk5')
        if is_login != []:
            break
    
    return
```

```python
def crawl_instagram(result, hashtag, count):
    url = 'https://www.instagram.com/explore/tags/' + hashtag + '/'
    wd = webdriver.Chrome('C:/Users/YJeong/CRWL_김연정/chromedriver_win32/chromedriver.exe')
    
    sleep(3)
    
    login_instagram(wd)
    print("LOGIN SUCCESS")
    sleep(2)
    
    wd.get(url)
    print("SEARCH HASHTAG : " + url)
    sleep(5)
       
    while True:
        infinite_scroll(wd)
        
        data = wd.page_source
        soupData = BeautifulSoup(data, 'html.parser')
        sleep(3)
        
        posts = wd.find_elements_by_css_selector('div.v1Nh3.kIKUG._bz0w')
        
        if len(posts) >= count:
            break
    print("LOAD POSTS %d" % len(posts)) 
    
    print("CRAWL POSTS")
    crawl_posts(result, wd, posts, count)
    
    return
```

```python
def crawl_posts(result, wd, posts, count):
    
    cnt = 0
    
    for post in posts:
        # 게시물 팝업 클릭
        post_a = post.find_element_by_tag_name('a')
        wd.execute_script("arguments[0].click();", post_a)
        sleep(5)
        
        data = wd.page_source
        soupData = BeautifulSoup(data, 'html.parser')
        sleep(2)
        
        # 답글 버튼 클릭
        try:
            reply_buttons = wd.find_elements_by_css_selector('div.Igw0E.IwRSH.eGOV_.ybXk5._4EzTm')

            # 모든 답글 버튼 클릭
            for reply_button in reply_buttons:
                reply_button.click()

            # 다시 페이지 불러오기
            data = wd.page_source
            soupData = BeautifulSoup(data, 'html.parser')
            sleep(2)
        except:
            print('No Reply')
            
        hashtags = []
        
        try:
            post_contents = soupData.select('div.C4VMK > span')
        
            # 게시물 내용과 댓글 확인
            for content in post_contents:
                only_hashtags = re.findall(r'#[가-힣a-zA-Z0-9,\\]+', content.text)
                
                # 해시태그 문자 삭제
                for tag in only_hashtags:
                    removed_tag = tag.replace('#', '')
                    hashtags.append(removed_tag)
        except:
            print('No Contents')
            
        try:
            like = soupData.select('div.Nm9Fw > button > span')[0].text
        except IndexError:
            try:
                like = soupData.select('div.Nm9Fw > button')[0].text[4:-1]
                if like == '':
                    like = '0'
            except:
                continue
                
        date = soupData.select('time._1o9PC.Nzb55')[0]['datetime'][:10]
        
        print(like, date, hashtags)
        
        for hashtag in hashtags:
            result.append([like] + [date] + [hashtag])
        
        close_button = wd.find_element_by_css_selector('div.Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG')
        wd.execute_script("arguments[0].click();", close_button)
        sleep(5)
        
        cnt+=1
        if cnt == count:
            break
        
    return
```

```python
def main():
    result = []
    search_hashtag = input("검색할 해시태그를 입력하세요 : ")
    search_count = input("검색할 게시물 수를 입력하세요 : ")
    
    print('---CRAWLING START---')
    
    crawl_instagram(result, str(search_hashtag), int(search_count))
    
    post_table = pd.DataFrame(result, columns=('like', 'date', 'hashtag'))
    post_table.to_csv("./instagram_hashtag_" + str(search_hashtag) + ".csv", encoding="cp949", mode='w', index=True)
    
    print('---CRAWLING FINISH---')
```

```python
main()
```

### 시각화 함수

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud, STOPWORDS
```

```python
def make_colors(word, font_size, position, orientation, random_state, **kwargs):
    r = random_state.randint(0, 255)
    g = random_state.randint(0, 255)
    b = random_state.randint(0, 255)
    
    color = "rgb(%d, %d, %d)" % (r, g, b)
    
    return color
```

```python
def make_word_cloud(hashtag_count, stopwords): 
    wc = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf',
                   stopwords=stopwords,
                   background_color='white', max_font_size=1000)
    generate = wc.generate_from_frequencies(hashtag_count)
    re_color_wc = generate.recolor(color_func=make_colors, random_state=True)

    plt.imshow(re_color_wc, interpolation='lanczos')
    plt.axis('off')
    plt.show()
    
    wc.to_file('hashtag.png')
    
    return
```

```python
def word_main():
    search_hashtag = ['커피맛집', '카페맛집', '카페투어']
    
    hashtags_table = pd.DataFrame([])
    
    for hashtag in search_hashtag:
        post_table = pd.read_csv("./instagram_hashtag_" + hashtag + ".csv", encoding="cp949")
        hashtags = post_table['hashtag']
        hashtags_table = pd.concat([hashtags_table, hashtags])
        
    hashtags_table.columns = ['hashtag']
    hashtags_pd = hashtags_table['hashtag']
    
    hashtag_count = Counter(hashtags_pd)
    
    stopwords = set(STOPWORDS)
    stopwords.add('카페')
    stopwords.add('커피맛집')
    stopwords.add('카페맛집')
    stopwords.add('카페투어')
    
    make_word_cloud(hashtag_count, stopwords)
```

```python
word_main()
```
